# Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.

```python
df = df[~(df.Confirmed < 10)]
```

```python
df.head()
```

# Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(5)
```

# Q. 4) In which Region, minimum number of Deaths cases were recorded ?

```python
df.groupby('Region')['Deaths'].sum().sort_values().head(50)
```

# Q. 5) How many Confirmed, Deaths & Recovered cases were reported from Pakistan till 29 April 2020 ?

```python
df[df['Region'].isin(["Pakistan"])]
```
