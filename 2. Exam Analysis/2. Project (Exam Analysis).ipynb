```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Exam Data/Expanded_Data_with_more_features.csv")
```

#### Deleting Unamed series from data frame


```python
df.drop('Unnamed: 0',axis=1,inplace = True)
```

### Gender Distributution



```python
custom_palette = ["red", "orange"]
ax = sns.countplot(data = df,x ='Gender',palette=custom_palette)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.annotate('Point (3, 7)', xy=(3, 7), xytext=(3.5, 8),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Analysis')
plt.show()
```

* Conclusion: Female count is slightly greater than males 

#### Does Parents marital status effects student studies?

```python
pm = df.groupby('ParentMaritalStatus').agg({'MathScore':"mean","ReadingScore":'mean',"WritingScore":'mean'})

sns.heatmap(pm,annot = True)
plt.show
```

* Conclusion: Parents Marital status does not effect on studies of their child

#### Does Parents Eduction has effects student studies?

```python
pm2 = df.groupby('ParentEduc').agg({'MathScore':"mean","ReadingScore":'mean',"WritingScore":'mean'})

sns.heatmap(pm2,annot = True)
plt.show
```

* Conclusion: Parents Education does not effect on studies of their child

# Distribution of Ethnic groups

```python
groupA = df.loc[(df['EthnicGroup']=='group A')].count()
groupB = df.loc[(df['EthnicGroup']=='group B')].count()
groupC = df.loc[(df['EthnicGroup']=='group C')].count()
groupD = df.loc[(df['EthnicGroup']=='group D')].count()
groupE = df.loc[(df['EthnicGroup']=='group E')].count()

l = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E']
groups = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(groups,labels= l,autopct = '%1.2f%%')
plt.show
```
